```python
import tushare as ts
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

    code：股票代码，即6位数字代码，或者指数代码（sh=上证指数 sz=深圳成指 hs300=沪深300指数 sz50=上证50 zxb=中小板 cyb=创业板）
    start：开始日期，格式YYYY-MM-DD
    end：结束日期，格式YYYY-MM-DD
    ktype：数据类型，D=日k线 W=周 M=月 5=5分钟 15=15分钟 30=30分钟 60=60分钟，默认为D
    retry_count：当网络异常后重试次数，默认为3
    pause:重试时停顿秒数，默认为0


```python
market_index=['sh','sz','hs300','sz50','zxb','cyb']
```

```python
container={}
for market_symbol in market_index:
    data=ts.get_hist_data(market_symbol,ktype='d',start='2000-10-01',end="2017-10-01")
    # data[:2]
    close_rate=(data.close-data.close.shift(-1))/data.close.shift(-1)
    plt.hist(close_rate.dropna(),bins=50)
    container[market_symbol]=[close_rate.mean(),close_rate.std()]
    print(' {} skewness ={} ;kurtosis = {} '.format(market_symbol,close_rate.skew(),close_rate.kurtosis()))


#     print("{:*^50}".format(market_symbol))
#     print('mean=   ',close_rate.mean())
#     print('std=   ',close_rate.std())
```

```python
container={}
for market_symbol in market_index:
    data=ts.get_hist_data(market_symbol,ktype='w',start='2000-10-01',end="2017-10-01")
    # data[:2]
    close_rate=(data.close-data.close.shift(-1))/data.close.shift(-1)
    plt.hist(close_rate.dropna(),bins=50)
    container[market_symbol]=[close_rate.mean(),close_rate.std()]
    print(' {} skewness ={} ;kurtosis = {} '.format(market_symbol,close_rate.skew(),close_rate.kurtosis()))


#     print("{:*^50}".format(market_symbol))
#     print('mean=   ',close_rate.mean())
#     print('std=   ',close_rate.std())
```

```python
index=[]
mean=[]
std=[]
for  element in  (container):
    index.append(element)
    mean.append(container[element][0])
    
    std.append(container[element][1])
    

x=pd.DataFrame({'mean':mean,'std':std},index=index)

x.plot(kind='bar',sharex=True)
```

```python
x.plot(kind='bar', subplots=2,sharex=True)
```
